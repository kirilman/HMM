```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
signals = np.load('signals.npy')
```

```python
signals.shape
```

```python
fig = plt.figure(dpi = 150, figsize=( 6, 2))
plt.plot(signals[2])
plt.plot(signals[0],'g')
plt.plot(signals[4],'r')
```

```python
fig,ax = plt.subplots(3,2,figsize=(8, 4),sharex=True, sharey = True)
# fig.figsize=(2, 10)
fig.dpi = 120

for i in range(3):
    ax[i,0].plot(signals[i])
    ax[i,0].set_label(str(i))
    ax[i,0].grid()
for i in range(3):
    ax[i,1].plot(signals[i+3],'r')
    ax[i,1].set_label(str(i))
    ax[i,1].grid()

plt.savefig('  _1.png')
```

```python
fig,ax = plt.subplots(3,2,figsize=(8, 4),sharex=True, sharey = True)
# fig.figsize=(2, 10)
fig.dpi = 120
ax[0,1].set_title = 'Train'
for i in range(3):
    ax[i,0].plot(signals[i])
    ax[i,0].set_label(str(i))
    ax[i,0].grid()
for i in range(3):
    ax[i,1].plot(signals[i+6],'r')
    ax[i,1].set_label(str(i))
    ax[i,1].grid()

plt.savefig('  .png')
```
