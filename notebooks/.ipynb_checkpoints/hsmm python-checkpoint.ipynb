```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import numpy as np
from autohmm import ar

model = ar.ARTHMM(n_unique=2, n_features=1,n_lags=1, transmat_init=np.array([[0.1, 0.9],
                                                                         [0.6, 0.4]]))
model.transmat_
model.mu_ = np.array([1.0, 5.0])
model.var_ = np.array([0.15, 0.05])

samples, states = model.sample(n_samples=1000)
```

```python
print(samples.shape)
print(model.transmat_)
```

```python
plt.plot(samples)
plt.plot(states)
```
